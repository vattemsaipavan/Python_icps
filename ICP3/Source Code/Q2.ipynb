### Write a simple program that parse a Wiki page

```python
from bs4 import BeautifulSoup
import requests
```

```python
def scrap(url):
#   we need to parse the url using libraries requests and bs4  
    soup = requests.get(url)
    page_soup = BeautifulSoup(soup.content, "html.parser")
    
#   Print the title of the page
    print("Title of the page: {}".format(page_soup.title.string))
    
#   Extracting the links from the url
    for link in page_soup.find_all('a'):
        with open("links.txt","a+") as file:
            file.write("Anchor Tags in the Page: {}\n".format(link.get('href')))
        print("Anchor Tags in the Page: {}".format(link.get('href')))
        
url = "https://en.wikipedia.org/wiki/Deep_learning"
scrap(url)
```
